```python
import os
import numpy as np
```

```python
face_data = []
face_labels = []
name = {}
cid = 0
for file in os.listdir('./data'):
    if file.endswith('.npy'):
        fd = np.load('./data/'+file)
        face_data.append(fd)
        n = file[:-4]
        name[cid] = n
        face_labels.append(np.ones(fd.shape[0])*cid)
        cid += 1
```

```python
face_data = np.concatenate(face_data)
face_labels = np.concatenate(face_labels)
```
