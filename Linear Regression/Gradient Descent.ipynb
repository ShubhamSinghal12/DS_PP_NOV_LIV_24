```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x**2 + 3*x+5
```

```python
X = np.linspace(-4,3,1000)
y = f(X)
```

```python
plt.plot(X,y)
plt.show()
```

```python
def det_fx(x):
    return 4*x+3
```

```python
alpha = 0.01
x = -3
epochs = 1000
xs = [x]
for e in range(epochs):
    x = x - alpha*det_fx(x)
    xs.append(x)
x
```

```python
plt.plot(X,y)
plt.plot(xs,f(np.array(xs)),c='r')
plt.show()
```

```python
x = np.array([2,3,6,7])
y = np.array([3,2,7,6])

w0,w1 = 1,1
alpha = 0.05
ypred = w1*x+w0
```

```python
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
plt.show()
```

```python
ypred = w1*x+w0
w1 = w1 - alpha*np.mean((ypred-y)*x)
w0 = w0 - alpha*np.mean(ypred-y)
```
