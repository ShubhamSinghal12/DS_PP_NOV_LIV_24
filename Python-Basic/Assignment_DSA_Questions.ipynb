### Boston Numbers

```python
def sumOfDigits(n):
    sum = 0
    while n > 0:
        sum += n%10
        n //= 10
    return sum

def primeFactorsBostonSum(n):
    div = 2
    sum = 0
    while n > 1:
        while n%div == 0:
            sum += sumOfDigits(div)
            n //= div
        div += 1
    return sum

def boston(n):
    return sumOfDigits(n) == primeFactorsBostonSum(n)

# primeFactors(378)
```

```python
for i in range(2,1000):
    if boston(i):
        print(i)
```
