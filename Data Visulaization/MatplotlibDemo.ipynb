````python
X = np.arange(1,11)
y = 2*X+3
plt.plot(X,y)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Line Plot")
plt.show()
````

````python
X = np.arange(-11,11,0.2)
y = X**3
plt.plot(X,y)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-8,8)
plt.ylim(-80,80)
plt.title("Line Plot")
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
plt.plot(X,y1,label = 'Y1')
plt.plot(X,y2,'-o',label = 'Y2')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Line Plot")
plt.legend(loc = 'upper right')
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }

plt.plot(X,y1,label = 'Y1',marker = '*',color='green',linewidth = 0.5,markersize = 10)
plt.plot(X,y2,label = 'Y2',marker = 'o',linestyle = '-.')
plt.grid()
plt.xlabel('X',fontdict=font)
plt.ylabel('Y',fontdict=font)
plt.title("Line Plot",fontdict=font)
plt.legend(loc = 'upper right')
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
plt.scatter(X,y1,label = 'Y1',s = 20,marker = '*',c='green',linewidths=2,edgecolors='r')
plt.scatter(X,y2,label = 'Y2',marker = 'o')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Line Plot")
plt.legend(loc = 'upper right')
plt.show()
````
