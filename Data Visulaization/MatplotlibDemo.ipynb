````python
X = np.arange(-11,11,0.2)
y = X**3
line = plt.plot(X,y,color='g')
plt.setp(line,color = 'r')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-8,8)
plt.ylim(-80,80)
plt.title("Line Plot")
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
plt.plot(X,y1,label = 'Y1')
plt.plot(X,y2,':o',label = 'Y2')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Line Plot")
plt.legend(loc = 'upper right')
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }

plt.plot(X,y1,label = 'Y1',marker = '*',color='green',linewidth = 0.5,markersize = 10)
plt.plot(X,y2,label = 'Y2',marker = 'o',linestyle = '-.')
plt.grid()
plt.xlabel('X',fontdict=font)
plt.ylabel('Y',fontdict=font)
plt.title("Line Plot",fontdict=font)
plt.legend(loc = 'upper right')
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
plt.scatter(X,y1,label = 'Y1',s = 20,marker = '*',c='green',linewidths=2,edgecolors='r')
plt.scatter(X,y2,label = 'Y2',marker = 'o')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Line Plot")
plt.legend(loc = 'upper right')
plt.show()
````

````python
X = np.arange(1,11)
y1 = 2*X+3
y2 = X**2 + X + 5
plt.plot(X,y1,'-.o',X,y2,'--^')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Line Plot")
# plt.legend()
plt.savefig('LinePlot.jpg')
plt.show()
````

````python
X = np.linspace(-10,10,1001)
ySin = np.sin(X)

plt.figure(figsize=(10,2))
plt.plot(X,ySin)
plt.grid()
plt.xlim((-7.5,7.5))
plt.title("Sin Wave")
plt.show()

yCos = np.cos(X)

plt.figure(figsize=(10,2))
plt.plot(X,yCos)
plt.grid()
plt.xlim((-7.5,7.5))
plt.title("Cos Wave")
plt.show()
````

````python
X = np.linspace(-1,1,1001)
ySin = np.sin(2*np.pi*X)
yCos = np.cos(2*np.pi*X)
ytan = np.tan(2*np.pi*X)
ycot = 1/ytan[ytan != 0]

plt.figure(figsize=(12,4))
plt.subplot(2,2,1)
plt.plot(X,ySin)
plt.grid()
# plt.xlim((-7.5,7.5))
plt.title("Sin Wave")

plt.subplot(2,2,2)
plt.plot(X,yCos)
plt.grid()
# plt.xlim((-7.5,7.5))
plt.title("Cos Wave")

plt.subplot(2,2,3)
plt.plot(X,ytan)
plt.grid()
# plt.xlim((-7.5,7.5))
plt.title("Tan Wave")

plt.subplot(2,2,4)
plt.plot(X[ytan != 0],ycot)
plt.grid()
# plt.xlim((-7.5,7.5))
plt.title("Cot Wave")

plt.suptitle("Trignometric Functions")
plt.subplots_adjust(hspace=0.4,wspace=0.1)
plt.show()
````

````python
X = np.array([10,20,30])
y2024 = np.array([70,80,100])
y2025 = np.array([80,93,95])

plt.bar(X-1,y2024,label = "2024",width=2)#,tick_label = ['Gold','Silver','Platinum'])
plt.bar(X+1,y2025,label = "2025",width=2)
plt.xticks(X,['Gold','Silver','Platinum'])

plt.legend()
plt.show()
````
